```python
def match_and_draw(kp1, kp2, des1, des2, grayImage1, grayImage2):
    raw_matches = flann.knnMatch(des1, trainDescriptors = des2, k = 2) #2
    return filter_matches(kp1, kp2, raw_matches)

def filter_matches(kp1, kp2, matches, ratio = 0.75):
    mkp1, mkp2 = [], []
    for m in matches:
        if len(m) == 2 and m[0].distance < m[1].distance * ratio:
            m = m[0]
            mkp1.append( kp1[m.queryIdx] )
            mkp2.append( kp2[m.trainIdx] )
    p1 = np.float32([kp.pt for kp in mkp1])
    p2 = np.float32([kp.pt for kp in mkp2])
    kp_pairs = zip(mkp1, mkp2)
    return len(kp_pairs)
```

```python
def color_transfer(source, target):
    source = cv2.cvtColor(source, cv2.COLOR_BGR2LAB).astype("float32")
    target = cv2.cvtColor(target, cv2.COLOR_BGR2LAB).astype("float32")
    (lMeanSrc, lStdSrc, aMeanSrc, aStdSrc, bMeanSrc, bStdSrc) = image_stats(source)
    (lMeanTar, lStdTar, aMeanTar, aStdTar, bMeanTar, bStdTar) = image_stats(target)

    (l, a, b) = cv2.split(target)
    l -= lMeanTar
    a -= aMeanTar
    b -= bMeanTar
    
    l = (lStdTar / lStdSrc) * l
    a = (aStdTar / aStdSrc) * a
    b = (bStdTar / bStdSrc) * b

    l += lMeanSrc
    a += aMeanSrc
    b += bMeanSrc
    
    l = np.clip(l, 0, 255)
    a = np.clip(a, 0, 255)
    b = np.clip(b, 0, 255)

    transfer = cv2.merge([l, a, b])
    transfer = cv2.cvtColor(transfer.astype("uint8"), cv2.COLOR_LAB2BGR)

    return transfer
```

```python
def image_stats(image):
	
	# compute the mean and standard deviation of each channel
	(l, a, b) = cv2.split(image)
	(lMean, lStd) = (l.mean(), l.std())
	(aMean, aStd) = (a.mean(), a.std())
	(bMean, bStd) = (b.mean(), b.std())

	# return the color statistics
	return (lMean, lStd, aMean, aStd, bMean, bStd)
```

```python
def whiteBalance(target, matched):
    #Calculating alpha coefficients
    targetf = target.astype('float')
    alpha=[]
    [rows, cols,r]=targetf.shape
    for k in range(0,3):
        top=0.0
        bttm=0.0
        for i in range(0,rows):
            for j in range(0, cols):
                top=top+targetf[i,j,k]*matched[i,j,k]
                bttm=bttm+targetf[i,j,k]*targetf[i,j,k]
        alpha.append(top/bttm)
        #print top/bttm
            
    for k in range(0,3):
        for i in range(0,rows):
            for j in range(0, cols):
                targetf[i,j,k]=targetf[i,j,k]*alpha[k]
                if targetf[i,j,k]>255.0:
                    targetf[i,j,k]=255.0

    target = targetf.astype('uint8')
    return target
```

```python
def exposure(target, source):
    targetf=target.astype('float')
    sourcef=source.astype('float')
    [rows, cols, r]=target.shape
    
    top=0.0
    bttm=0.0
    for k in range(0,3):
        for i in range(0,10):
            for j in range(0, 10):
                top=top+targetf[i,j,k]*sourcef[i,j,k]
                bttm=bttm+targetf[i,j,k]*targetf[i,j,k]
                
    alpha=top/bttm
    #print alpha
    for k in range(0,3):
        for i in range(0,10):
            for j in range(0, 10):
                #print targetf[i,j,k]
                targetf[i,j,k]=target[i,j,k]*alpha
                #print targetf[i,j,k]
                if targetf[i,j,k]>255.0:
                    targetf[i,j,k]=255.0
                if targetf[i,j,k]<0.0:
                    targetf[i,j,k]=0.0
                    
    target=targetf.astype('uint8')
    return target
```

```python
def contrast(target, source):
    targetf=target.astype('float')
    sourcef=source.astype('float')
    [rows, cols, r]=target.shape
    for k in range(0,3):
        for i in range(0,rows):
            for j in range(0, cols):
                if targetf[i,j,k]==0.0:
                    targetf[i,j,k]=1.0
                if sourcef[i,j,k]==0.0:
                    sourcef[i,j,k]=1.0
    top=0.0
    bttm=0.0
    tar=log(targetf)
    src=log(sourcef)
    
    for k in range(0,3):
        for i in range(0,10):
            for j in range(0, 10):
                top=top+targetf[i,j,k]*targetf[i,j,k]*tar[i,j,k]*src[i,j,k]
                
                bttm=bttm+targetf[i,j,k]*targetf[i,j,k]*tar[i,j,k]*tar[i,j,k]
                
    alpha=top/bttm
    #print alpha
    for k in range(0,3):
        for i in range(0,10):
            for j in range(0, 10):
                #print targetf[i,j,k]
                targetf[i,j,k]=pow(target[i,j,k],alpha)
                #print targetf[i,j,k]
                if targetf[i,j,k]>255.0:
                    targetf[i,j,k]=255.0
                if targetf[i,j,k]<0.0:
                    targetf[i,j,k]=0.0
                    
    target=targetf.astype('uint8')
    return target
```

```python
def brightness(image, beta):
    imagef = image.astype('float')
    imagef = imagef + beta 
    [rows,cols,c] = imagef.shape
    for c in range(0,3):
        for i in range(0,rows):
            for j in range (0,cols):
                if imagef[i,j,c] > 255:
                    imagef[i,j,c] = 255
                if imagef[i,j,c] < 0:
                    imagef[i,j,c] = 0
 
    imhance = imagef.astype('uint8')        
    return imhance
```
